```python
#remove consicutive dublicate elements

def remove_dublicate(word):
    i=0
    li=list(word)
    size=len(li)
    j=1
    while(j<size):
        if li[j]!=li[i]:
            i+=1
            li[i]=li[j]
            j+=1
            
        else:
            j+=1
    srin=""
    List=li[:i+1]
    
    for i in List:
        srin=srin+i
    return srin



def Find_meaning(word):
    new_word=[]
    word=word.lower()
    count=0
    #removing the front spaces
    for i in word:
            if i==" ":
                count+=1
                
    word=word[count:]
    
    if word in data:
        if type(data[word])==list:
            
            for words in data[word]:
                print("-->",words)
                
        else:
            print(data["words"])
        
    elif word.title() in data:
        if type(data[word.title()])==list:
            
            for words in data[word.title()]:
                print("-->",words)
                
        else:
            print(data[word.title()])
            
    elif word.upper() in data:
        if type(data[word.upper()])==list:
            
            for words in data[word.upper()]:
                print("-->",words)
                
        else:
            print(data[word.upper()])
        
            
    else:
        print("word not found !!!")
        
        wor=word[:len(word)-1]
        wor2=word[1:]
        
        wor3=remove_dublicate(word)
        
        print("suggested words are ")
        if wor in data:
            print(wor)
        if wor2 in data:
            print(wor2)
        if wor3 in data:
            print(wor3)
            
    

word=input("Enter the word for which you want to know meaning ")


Find_meaning(word)
```
